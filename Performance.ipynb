```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['CNN', 'InceptionV3', 'VGG19', 'ResNet152']
resolutions = ['90x120', '360x480', 'Latent']

accuracy_data = np.array([
    [87.70, 84.49, 77.01, 68.98],
    [91.98, 97.33, 99.47, 78.07],
    [94.65, 93.05, 94.65, 79.14]
])

specificity_data = np.array([
    [0.96, 0.95, 0.93, 0.9],
    [0.97, 0.99, 1, 0.92],
    [0.98, 0.98, 0.98, 0.93]
])

training_time_data = np.array([
    [30.76, 96.39, 65.22, 286.48],
    [149.6, 408.13, 581.38, 1757.25],
    [44.02, 82.5, 88.77, 293.23]
])

# Function to generate colors based on resolution
def generate_colors(resolution):
    if resolution == '90x120':
        return 'lightgrey'
    elif resolution == '360x480':
        return 'grey'
    elif resolution == 'Latent':
        return 'darkgrey'

# Function to add labels above bars
def add_labels(bars, data, multiplier=1, height_offset=1):
    for bar, value in zip(bars, data):
        plt.text(bar.get_x() + bar.get_width() / 2 - 0.05, bar.get_height() + height_offset * multiplier, f'{value:.2f}', ha='center', va='bottom')

# Plotting Accuracy
plt.figure(figsize=(15, 5))

for i, resolution in enumerate(resolutions):
    bars = plt.bar(np.arange(len(models)) + i * 0.25, accuracy_data[i], width=0.25, label=resolution, color=generate_colors(resolution))
    add_labels(bars, accuracy_data[i])

plt.xlabel('Models')
plt.ylabel('Accuracy (%)')
plt.title('Accuracy Comparison Across Models and Resolutions')
plt.xticks(np.arange(len(models)) + 0.25, models)
plt.legend(title='Resolution')
plt.ylim(0, 120)
plt.show()

# Plotting Specificity
plt.figure(figsize=(15, 5))

for i, resolution in enumerate(resolutions):
    bars = plt.bar(np.arange(len(models)) + i * 0.25, specificity_data[i], width=0.25, label=resolution, color=generate_colors(resolution))
    add_labels(bars, specificity_data[i], multiplier=0.1, height_offset=0.05)

plt.xlabel('Models')
plt.ylabel('Specificity')
plt.title('Specificity Comparison Across Models and Resolutions')
plt.xticks(np.arange(len(models)) + 0.25, models)
plt.legend(title='Resolution')
plt.ylim(0, 1.4)
plt.show()

# Plotting Training Time
plt.figure(figsize=(15, 5))

for i, resolution in enumerate(resolutions):
    bars = plt.bar(np.arange(len(models)) + i * 0.25, training_time_data[i], width=0.25, label=resolution, color=generate_colors(resolution))
    add_labels(bars, training_time_data[i], multiplier=50)

plt.xlabel('Models')
plt.ylabel('Training Time (s)')
plt.title('Training Time Comparison Across Models and Resolutions')
plt.xticks(np.arange(len(models)) + 0.25, models)
plt.legend(title='Resolution')
plt.ylim(0, 1900)
plt.show()
```
